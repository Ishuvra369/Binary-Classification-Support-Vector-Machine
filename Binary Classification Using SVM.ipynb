# Binary Classification Using SVM

```python
# Importing the Libraries

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the Dataset

dataset = pd.read_csv("E:\Python for Data Science and Machine Learning\Social_Network_Ads.csv")
pd.set_option('display.max_columns', None)
dataset.head()
```

```python
# Segregating Independent Variable

X = dataset.iloc[:,[1,2] ]
y = dataset.iloc[:,[3] ]
```

```python
# Splitting the Dataset into the Training set and Test set

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
# Feature Scaling

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Linear Kernel 

from sklearn.svm import SVC
```

```python
# Running the model with "linear"

classifier = SVC(kernel = "linear", random_state = 0,gamma='auto')
classifier.fit(X_train, y_train)
```

```python
# Predicting the Test set results

y_pred = classifier.predict(X_test)
```

```python
y_pred
```

```python
# Validation of the model

from sklearn import metrics
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred))
```

```python
# Full Report

print(classification_report(y_test, y_pred))
```

```python
# Fitting SVM to the Training set

from sklearn.svm import SVC
```

```python
all_kernel = ['linear', 'poly', 'rbf', 'sigmoid'] 

for kernel in all_kernel: 

    classifier = SVC(kernel = kernel, random_state = 0,gamma='auto')
    classifier.fit(X_train, y_train)
    # Predicting the Test set results
    y_pred = classifier.predict(X_test)
    
    from sklearn import metrics
    from sklearn.metrics import classification_report, confusion_matrix
    acc = metrics.accuracy_score(y_test, y_pred)
    print("For kernel ", kernel, " accuracy is ", acc)
```

```python
# Running the model with "rbf"

classifier = SVC(kernel = "rbf", random_state = 0,gamma='auto')
classifier.fit(X_train, y_train)
```

```python
# Predicting the Test set results

y_pred = classifier.predict(X_test)
```

```python
y_pred
```

```python
# Validation of the Model

from sklearn import metrics
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred))
```

```python
# Full Report

print(classification_report(y_test, y_pred))
```

```python
# Visualising the Training set results

from matplotlib.colors import ListedColormap
X_set, y_set = X_train, y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
plt.show()
```

```python
# Visualising the Test set results

from matplotlib.colors import ListedColormap
X_set, y_set = X_test, y_test
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
plt.show()
```
